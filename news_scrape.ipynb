```python
import time
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def scrape_gatech_news(base_url):
    # Setup undetected ChromeDriver
    options = uc.ChromeOptions()
    options.add_argument('--headless')  # Run headless if you don't need a browser UI
    driver = uc.Chrome(options=options)
    
    # Start URL
    driver.get(base_url)
    
    # Wait for the page to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, ".views-element-container"))
    )
    
    try:
        while True:
            # Extract articles
            articles = driver.find_elements(By.CSS_SELECTOR, '.views-row .h3 a')
            for article in articles:
                category = str(driver.current_url).split("?")[0].split("/")[-1].replace("-","_")
                print(category)
                url = article.get_attribute('href')
                print(url)

                url_dict.update({url: category})
            
            # Try to find and click the 'Next' pagination link
            next_page_link = driver.find_element(By.XPATH, '//*[@title="Go to next page"]')
            if next_page_link:
                driver.execute_script("arguments[0].click();", next_page_link)
                # next_page_link.click()
                # time.sleep(2)  # Wait for the page to load
            else:
                break  # If no 'Next' link, exit the loop
    except Exception as e:
        print("Error or last page reached:", e)
    finally:
        driver.quit()

if __name__ == '__main__':
    url_roots = ['https://news.gatech.edu/categories/society-and-culture?page=0',
             'https://news.gatech.edu/categories/science-and-technology?page=0',
             'https://news.gatech.edu/categories/health-and-medicine?page=0',
             'https://news.gatech.edu/categories/earth-and-environment?page=0',
             'https://news.gatech.edu/categories/campus-and-community?page=0',
             'https://news.gatech.edu/categories/business-and-economic-development?page=0']
    for base_url in url_roots:
        scrape_gatech_news(base_url)
```
